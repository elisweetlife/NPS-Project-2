```python
designation = []
fullName = []
name = []
parkCode = []
states = []
url = []
```

```python
for park in park_list:
    designation.append(park["designation"])
    fullName.append(park["fullName"])
    name.append(park["name"])
    parkCode.append(park["parkCode"])
    states.append(park["states"])
    url.append(park["url"])

```
